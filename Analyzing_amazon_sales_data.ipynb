```python
#Importing all Libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Importing dataset.
df = pd.read_csv("amazon.csv")
df
```

```python
#Using .describe() command to get the overview of the dataframe.
df.describe()
```

```python
df.info()
```

```python
#Since there are currnecy symbols, commas, percentage symbols, these symbols needs to be removed in order to change the data type.
#.str.replace() command is used to remove currency symbol and commas from the columns actual_price and discounted_price.
#.astype() command is used to convert the data type in these columns to float64.
df['discounted_price'] = df['discounted_price'].str.replace("‚Çπ",'').str.replace(",",'').astype('float64')
df['actual_price'] = df['actual_price'].str.replace("‚Çπ",'').str.replace(",",'').astype('float64')
```

```python
df['discount_amount'] = df['actual_price'] - df['discounted_price']
```

```python
#.str.replace() command is used to remove percentage symbol from discount_percentage column.
#..astype() command is used to convert the data type in this column to float64.
df['discount_percentage'] = df['discount_percentage'].str.replace('%','').astype('float64')
df['discount_percentage'] = df['discount_percentage'] / 100
```

```python
df['discount_percentage']
```

```python
#.value_counts() command is used to display the unique values in a column and also the number of each unique values.
df['rating'].value_counts()
```

```python
#.loc command is used to seelct the row where the rating value is |.
df.loc[df['rating'] == '|']
```

```python
#using the .str.replace() and .astype() commands to replace | with 3.9 (assuming average rating for the product is 3.9) and coverting
#the data type to float64 in the rating column.
df['rating'] = df['rating'].astype(str)
df['rating'] = df['rating'].str.replace('|', '3.9')
df['rating'] = df['rating'].astype('float64')
```

```python
#Using the .str.replace() and .astype() commands to remove the commas and convert the datatype to float64 in the rating_count column.
df['rating_count'] = df['rating_count'].str.replace(',', '').astype('float64')
```

```python
#using .isna() and .sum() command to observe the number of null values in each column.
df.isna().sum()
```

```python
#using the .index command and .iloc command to find the index number of rows with 1 or more null values and displaying the rows.
df.iloc[df[(df.isna().sum(axis = 1) >= 1)].index]
```

```python
#To calculate the average rating_count using the .mean() command, using the round() command to round the average to nearest whole
#number and using the .fillna() command to replace null values with the rounded average value.
df['rating_count'] = df['rating_count'].fillna(round(df["rating_count"].mean()))
```

```python
df.isna().sum()
```

```python
#.drop_duplicates() command is used to remove the duplicate rows.
df = df.drop_duplicates()
```

```python
#.str.strip() command is used to remove the whiespaces or blank spaces at the begining or end of the string.
df['product_id'].str.strip()
```

```python
#In order to split or break the string values in the category column using the .str.split() command with | as a delimiter.
#Every split string element is inserted into separate column using expand = True.
breakcategory = df['category'].str.split('|', expand = True)
breakcategory
```

```python
breakcategory.isna().sum()
```

```python
#.rename() command is used to rename the column 0 and column 1 to category and subcategory.
breakcategory = breakcategory.rename(columns = {0:'category', 1:'subcategory'})
breakcategory
```

```python
#.unique() command is used to display the unique values in the category column.
breakcategory['category'].unique()
```

```python
breakcategory['category'] = breakcategory['category'].str.replace('&', ' & ')
breakcategory['category'] = breakcategory['category'].str.replace('MusicalInstruments', 'Musical Instruments')
breakcategory['category'] = breakcategory['category'].str.replace('OfficeProducts', 'Office Products')
breakcategory['category'] = breakcategory['category'].str.replace('HomeImprovement', 'Home Improvement')
```

```python
breakcategory['subcategory'].unique()
```

```python
breakcategory['subcategory'] = breakcategory['subcategory'].str.replace('&', ' & ')
breakcategory['subcategory'] = breakcategory['subcategory'].str.replace(',', ', ')
breakcategory['subcategory'] = breakcategory['subcategory'].str.replace('NetworkingDevices', 'Networking Devices')
breakcategory['subcategory'] = breakcategory['subcategory'].str.replace('HomeTheater', 'Home Theater')
breakcategory['subcategory'] = breakcategory['subcategory'].str.replace('HomeAudio', 'Home Audio')
breakcategory['subcategory'] = breakcategory['subcategory'].str.replace('WearableTechnology', 'Wearable Technology')
breakcategory['subcategory'] = breakcategory['subcategory'].str.replace('ExternalDevices', 'External Devices')
breakcategory['subcategory'] = breakcategory['subcategory'].str.replace('DataStorage', 'Data Storage')
breakcategory['subcategory'] = breakcategory['subcategory'].str.replace('GeneralPurposeBatteries', 'General Purpose Batteries')
breakcategory['subcategory'] = breakcategory['subcategory'].str.replace('BatteryChargers', 'Battery Chargers')
breakcategory['subcategory'] = breakcategory['subcategory'].str.replace('OfficePaperProducts', 'Office Paper Products')
breakcategory['subcategory'] = breakcategory['subcategory'].str.replace('CraftMaterials', 'Craft Materials')
breakcategory['subcategory'] = breakcategory['subcategory'].str.replace('OfficeElectronics', 'Office Electronics')
breakcategory['subcategory'] = breakcategory['subcategory'].str.replace('PowerAccessories', 'Power Accessories')
breakcategory['subcategory'] = breakcategory['subcategory'].str.replace('HomeAppliances', 'Home Appliances')
breakcategory['subcategory'] = breakcategory['subcategory'].str.replace('AirQuality', 'Air Quality')
breakcategory['subcategory'] = breakcategory['subcategory'].str.replace('HomeStorage', 'Home Storage')
breakcategory['subcategory'] = breakcategory['subcategory'].str.replace('CarAccessories', 'Car Accessories')
breakcategory['subcategory'] = breakcategory['subcategory'].str.replace('HomeMedicalSupplies', 'Home Medical Supplies')
```

```python
#using the .drop() command to remove the category column from the dataframe. And adding the new category and subcategory column
#to the dataframe.
df = df.drop(columns = 'category')
df['category'] = breakcategory['category']
df['subcategory'] = breakcategory['subcategory']
```

```python
#In order to categorize rating scale a for and if else loop is created. .append() command is used to add rank categories 
#to a ranking list.
ranking = []
for score in df['rating']:
    if score <= .9 : ranking.append('Very Poor')
    elif score <= 1.9 : ranking.append('Poor')
    elif score <= 2.9 : ranking.append('Average')
    elif score <= 3.9 : ranking.append('Good')
    elif score <= 4.9 : ranking.append('Very Good')
    elif score == 5.0 : ranking.append('Excellent')
df['ranking'] = ranking
df['ranking'] = df['ranking'].astype('category')
```

```python
newdf = df[['product_id', 'product_name', 'actual_price', 'discounted_price', 'discount_amount', 'discount_percentage', 
                       'category', 'subcategory', 'rating', 'rating_count', 'ranking']]
newdf
```

```python
#In order to split each element of a list into a row .explode() command is used.
id_rows = split_user_id.explode()
name_rows = split_user_name.explode()
```

```python
#The below aommands are used to create testing and training data, perform linear regression using these data.
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
from sklearn.model_selection import cross_val_score, KFold
```

```python
X = correldf.iloc[:, 0].values.reshape(-1, 1)
Y = correldf.iloc[:, 1].values.reshape(-1, 1)
```

```python
#Using the train_test_split() command to create testing and training data. test_size = 0.2 tells the train_test_split() command to take 20% of the original dataset and use it as test data.
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2, random_state = 42)
```

```python
#Using the LinearRegression().fit() command to perform linear regression using the training data, .coef_ attribute to calcualte the coefficient, # Use the .intercept_ attribute to cacluate the intercept.
fitted = LinearRegression().fit(X_train, Y_train)
print('coefficient = ', fitted.coef_)
print('intercept = ', fitted.intercept_)
```

```python
#Using the .predict() to predict the values of the depdentant variable using the model, r2_score() command to calculate the R2 score by comparing the predicted dependant variable values with the actual dependant variable values.
Y_train_predict = fitted.predict(X_train)
print('R2 = ', r2_score(Y_train, Y_train_predict))
```

```python
# Using the .scatter() command to create a scatter plot using the training data, .plot() to create a regression line using the predicted Y data rather than the training Y data, and xlabel(), ylabel(), and title() commands to add labels to the scatter plot.
plt.scatter(X_train, Y_train)
plt.plot(X_train, Y_train_predict, color = 'black')
plt.xlabel('Actual Price')
plt.ylabel('Discounted Price')
plt.title('Actual Price vs. Discounted Price Regression (Training Data)')
plt.show()
```

```python
Y_test_predict = fitted.predict(X_test)
print('R2 = ', r2_score(Y_test, Y_test_predict))
```

```python
plt.scatter(X_test, Y_test)
plt.plot(X_test, Y_test_predict, color = 'black')
plt.xlabel('Actual Price')
plt.ylabel('Discounted Price')
plt.title('Actual Price vs. Discounted Price Regression (Test Data)')
plt.show()
```

```python
#Uisng the ols() command to perform ordinary least squared linear regresson, .predict command to predict the values of the dependant variable (discount price), .resid command to calculate the difference between the predicted values and actual values.
from statsmodels.formula.api import ols
model = ols(formula = 'discounted_price ~ actual_price', 
          data = correldf[['discounted_price', 'actual_price']]).fit()
X_OLS = correldf['actual_price']
fitted_values = model.predict(X_OLS)
residuals = model.resid
```

```python
plt.scatter(fitted_values, residuals)
plt.xlabel('Fitted Values')
plt.ylabel('Residuals')
plt.show()
```
